```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_boston
boston = load_boston()

feature_n=0,2 # selection of target features 選參數編號
x=boston.data[:,feature_n] # selection of target features
y=boston.target # extract house price as target label
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

from sklearn.svm import SVR
regressor = SVR(kernel='rbf', C=1e3, gamma=0.1)
regressor.fit(x_train, y_train)

y_predict=regressor.predict(x_test)

score = regressor.score(x_test,y_test)
print('score of regressor = ',score)
from sklearn.metrics import mean_squared_error
import numpy as np
MSE=mean_squared_error(y_test, y_predict)
RMSE=np.sqrt(MSE)
print('RMSE =' , RMSE)

import matplotlib.pyplot as plt
plt.figure()
plt.scatter(x[:,0],y)
# label = str(0)+":"+boston.feature_names[0]
plt.xlabel("Feature No.{:d} : {:s}".format(1,boston.feature_names[1]))
plt.ylabel('Median value of owner-occupied homes in 1000s')
plt.title('Relation between target and feature No. {:d} '.format(1))
plt.show()

```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_boston
boston = load_boston()
for i in range(13):
    feature_n=0,i # selection of target features 選參數編號
    x=boston.data[:,feature_n] # selection of target features
    y=boston.target # extract house price as target label
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

    from sklearn.svm import SVR
    regressor = SVR(kernel='rbf', C=1e3, gamma=0.1)
    regressor.fit(x_train, y_train)

    y_predict=regressor.predict(x_test)

    score = regressor.score(x_test,y_test)
    print('score of regressor = ',score)
    from sklearn.metrics import mean_squared_error
    import numpy as np
    MSE=mean_squared_error(y_test, y_predict)
    RMSE=np.sqrt(MSE)
    print('RMSE =' , RMSE)

    import matplotlib.pyplot as plt
    plt.figure()
    plt.scatter(x[:,1],y)
    # label = str(0)+":"+boston.feature_names[0]
    plt.xlabel("Feature No.{:d} : {:s}".format(i,boston.feature_names[i]))
    plt.ylabel('Median value of owner-occupied homes in 1000s')
    plt.title('Relation between target and feature No. {:d} '.format(i))
    plt.show()

```
