```python
from sklearn.datasets import load_wine
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import train_test_split

wine = load_wine()
data = wine.data
target = wine.target
print(wine.target_names)
print(wine.feature_names)

X_train,X_test,Y_train,Y_test = train_test_split(data,target)

classifier = tree.DecisionTreeClassifier(splitter='random',max_depth=3)

wine_clf =classifier.fit(X_train,Y_train)

predict_result = wine_clf.predict(X_test)

accuracy = accuracy_score(Y_test,predict_result)
print("Accuracy(正確率) ={:8.3f}%".format(accuracy*100))

plt.figure(figsize=(12,9))

# Decision Tree 特殊方法
tree.plot_tree(wine_clf,filled=True,class_names=wine.target_names,feature_names=wine.feature_names)
plt.show()
```
